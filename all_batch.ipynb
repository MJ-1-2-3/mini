```python
import pandas as pd
import numpy as np
import math 
batch = []
batch2 = []
count_of_students = 0
selected_class = []
roll_list = []
#Code for reading the batches for which exam is to be conducted

while(1):
  chk =int (input('Do you want to enter more batches'))
  if (chk == 1):
    a = input('Enter the batch for exam')
    batch.append(a)
    batch2.append(a)
  else:
    break
print(batch)


#Code for reading the csv files
n = 0
while n < len(batch):
  i = batch[n]
  if i == 'CS' :
    cs = pd.read_csv('CS Batch csv.csv')
    cs_roll = cs.iloc[:, 1]
    count_of_students += len(cs_roll)
    roll_list.append(cs_roll)
  elif i == 'CH' :
    ch = pd.read_csv('CH csv.csv')
    ch_roll = ch.iloc[:, 1]
    count_of_students += len(ch_roll)
    roll_list.append(ch_roll)
  elif i == 'CIV A' :
    civ_a = pd.read_csv('CIVIL A batch csv.csv')
    civ_a_roll = civ_a.iloc[:, 1]
    count_of_students += len(civ_a_roll)
    roll_list.append(civ_a_roll)
  elif i == 'CIV B' :
    civ_b = pd.read_csv('CIVIL B Batch csv.csv')
    civ_b_roll = civ_b.iloc[:, 1]
    count_of_students += len(civ_b_roll)
    roll_list.append(civ_b_roll)
  elif i == 'EC A' :
    ec_a = pd.read_csv('EC A Batch csv.csv')
    ec_a_roll = ec_a.iloc[:, 1]
    count_of_students += len(ec_a_roll)
    roll_list.append(ec_a_roll)
  elif i == 'EC B' :
    ec_b = pd.read_csv('EC B Batch csv.csv')
    ec_b_roll = ec_b.iloc[:, 1]
    count_of_students += len(ec_b_roll)
    roll_list.append(ec_b_roll)
  elif i == 'EEE A' :
    ee_a = pd.read_csv('EEE A Batch csv.csv')
    ee_a_roll = ee_a.iloc[:, 1]
    count_of_students += len(ee_a_roll)
    roll_list.append(ee_a_roll)
  elif i == 'EEE B' :
    ee_b = pd.read_csv('EEE B Batch csv.csv')
    ee_b_roll = ee_b.iloc[:, 1]
    count_of_students += len(ee_b_roll)
    roll_list.append(ee_b_roll)
  elif i == 'MECH A' :
    mc_a = pd.read_csv('MECH A Batch csv.csv')
    mc_a_roll = mc_a.iloc[:, 1]
    count_of_students += len(mc_a_roll)
    roll_list.append(mc_a_roll)
  elif i == 'MECH B' :
    mc_b = pd.read_csv('MECH B Batch csv.csv')
    mc_b_roll = mc_b.iloc[:, 1]
    count_of_students += len(mc_b_roll)
    roll_list.append(mc_b_roll)
  else:
    print("Invalid batch added")
  n = n + 1   

no_of_class =  math.ceil(count_of_students/45)
x = no_of_class


print(f'Total number of students :{count_of_students} ')
print(f'Total number of classes : {no_of_class}')

class_list = []
class Section:
  def __init__(self,name,rem_seat = 0):
    self.name = name
    self.alloted_seat = np.full((5,9),'NULL',dtype = 'U9')
    self.rem_seat = rem_seat

while x:
  y = input('Enter the class to be alloted')
  name = selected_class.append(y)
  c1 = Section(name)
  class_list.append(c1)
  x = x - 1
  print(c1.alloted_seat)
print(selected_class)
print(class_list)

c1 = class_list[0]
c1.alloted_seat[0][0] ="B20CSB42"
print(f'The value of first seat is : {c1.alloted_seat[0][0]}')

def allotment():
  print('The classes for allotment are : ')
  print(selected_class)
  print('The batches to be alloted are: ')
  print(batch)

  #Allotment of 2 batches at a time


  first = 0
  second = 1
  num_class = 0
  batch_num = 0
  while(len(batch2)):
    while num_class < len(selected_class):
        print('\n')
    
        # b1 = batch[first]
        # b2 = batch[second]
        c1 = class_list[num_class]

        j = 0
        k = 0
        count = 0
        if (len(roll_list[first]) + len(roll_list[second])) < 45:
            del batch2[0:2]
            print(batch2)
            #print('Inside if statement')
            break
            # print(roll_list[first])
            # print(roll_list[second])
        
        elif (len(roll_list[first]) < 25):
           first , second = second, first

        while k < 9:
            while j < 5:
                c1.alloted_seat[j][k] = roll_list[first].iloc[count]
                #print(c1.alloted_seat[j][k])
                j = j + 1
                roll_list[first] = roll_list[first].iloc[1:]

            j = 0
            k = k + 2
            #print('End of first while')


        l = 0
        m = 1
        count = 0

        while m < 9:
            while l < 5:
                #print('Inside second while')
                c1.alloted_seat[l][m] = roll_list[second].iloc[count]
                #print(c1.alloted_seat[l][m])
                l = l + 1
                roll_list[second] = roll_list[second].iloc[1:]

            l = 0
            m = m + 2
            #print('End of second while')

        i = 0
        j=0
        while i< 5 :
            while j<9 :
                #print('Inside third while')
                print(c1.alloted_seat[i][j],end = ' ')
                j = j + 1
            print('\n')
            j = 0
            i = i + 1

        num_class += 1
    batch_num += 1
    second += 2
    first += 2
    
    
        
print('Outside outermost while')

  
allotment()

#Have to consider the batches and the classes at the same time
#batch -> list of batches
#class_list -> list of class objects
#roll_list -> list containing roll numbers in the order of classes in class_list
```
